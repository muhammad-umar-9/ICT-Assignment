```python

import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas as pd

plt.style.use('seaborn-whitegrid')
data = {
    'Year': ['2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022', '2023', '2024'],
    'Innovation': [
        '4G LTE Expansion',
        'Blockchain Technology Emerges',
        'AI in Everyday Applications',
        '5G Networks Roll Out',
        'Growth of Cloud Computing',
        'Video Conferencing Platforms Take Off',
        'Surge in AR and VR Adoption',
        'Rise of Edge Computing',
        'Generative AI Becomes Mainstream',
        'Quantum Computing Emerges'
    ],
    'Impact': [6, 8, 9, 10, 7, 9, 8, 7, 9, 10]
}


df = pd.DataFrame(data)

df['Year'] = pd.to_datetime(df['Year'])

fig, ax = plt.subplots(figsize=(14, 9), dpi=100)

ax.bar(df['Year'], df['Impact'], color='skyblue', width=100, alpha=0.6, label='Impact Level')

ax.plot(df['Year'], df['Impact'], linestyle='-', marker='o', color='royalblue')

for i, row in df.iterrows():
    ax.text(row['Year'], row['Impact'] - 0.6, row['Innovation'], rotation=45, ha='center', fontsize=10, color='black')

ax.set_ylabel('Innovation Impact Level (1-10)', fontsize=12)
ax.set_ylim(0, 12)

ax.xaxis.set_major_locator(mdates.YearLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
plt.xticks(rotation=45)

plt.title('Timeline of ICT Innovations (2015-2024) with Impact Comparison', fontsize=16)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas as pd
from datetime import datetime
from matplotlib import cm
plt.style.use('seaborn-whitegrid')


data = {
    'Year': ['2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022', '2023', '2024'],
    'Innovation': [
        '4G LTE Expansion',
        'Blockchain Technology Emerges',
        'AI in Everyday Applications',
        '5G Networks Roll Out',
        'Growth of Cloud Computing',
        'Video Conferencing Platforms Take Off',
        'Surge in AR and VR Adoption',
        'Rise of Edge Computing',
        'Generative AI Becomes Mainstream',
        'Quantum Computing Emerges'
    ],
    'Impact Score': [3, 5, 7, 6, 8, 9, 6, 7, 8, 10]
}


df = pd.DataFrame(data)


df['Year'] = pd.to_datetime(df['Year'])

# Create figure
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 14), dpi=100)


cmap = cm.get_cmap('tab10')
ax1.plot(df['Year'], [1]*len(df), linestyle='-', marker='o', color=cmap(0), label='Innovation')


for i, row in df.iterrows():
    ax1.text(row['Year'], 0.98, row['Innovation'], rotation=45, ha='right', fontsize=10, color=cmap(1))

ax1.xaxis.set_major_locator(mdates.YearLocator())
ax1.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
plt.setp(ax1.get_xticklabels(), rotation=0)


ax1.set_xlabel('Year', fontsize=12)


ax1.set_ylabel('Timeline Position (Fixed)', fontsize=12)

ax1.set_title('Timeline of ICT Innovations (2015-2024)', fontsize=16)
ax1.grid(axis='y', linestyle='--', alpha=0.7)

ax2.bar(df['Year'].dt.year, df['Impact Score'], color=cmap(2))


for i, row in df.iterrows():
    ax2.text(row['Year'].year, row['Impact Score'] + 0.1, f'{row["Innovation"]}\n({row["Impact Score"]})',
             ha='center', fontsize=9, color=cmap(3))

ax2.xaxis.set_major_locator(mdates.YearLocator())
ax2.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
ax2.set_xlabel('Year', fontsize=12)


ax2.set_ylabel('Impact Score', fontsize=12)


ax2.set_title('Impact Score Comparison of ICT Innovations', fontsize=16)
ax2.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


years = np.array([2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022])
diagnosis_time = np.array([45, 42, 38, 35, 33, 28, 25, 20])
accuracy = np.array([75, 77, 79, 82, 85, 88, 90, 93])

fig, ax1 = plt.subplots(figsize=(10, 6))

bar_width = 0.4
ax1.bar(years - bar_width / 2, diagnosis_time, bar_width, color='cornflowerblue', label='Diagnosis Time (minutes)')
ax1.set_xlabel('Year')
ax1.set_ylabel('Diagnosis Time (minutes)', color='cornflowerblue')
ax1.tick_params(axis='y', labelcolor='cornflowerblue')


plt.title('Impact of AI-Based Diagnostic Tool on Healthcare Outcomes')


ax2 = ax1.twinx()
ax2.plot(years, accuracy, color='darkorange', marker='o', label='Accuracy (%)', linewidth=2)
ax2.set_ylabel('Accuracy (%)', color='darkorange')
ax2.tick_params(axis='y', labelcolor='darkorange')

ax1.legend(loc='upper left')
ax2.legend(loc='upper right')


plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()


plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
data = {
    'Year': [2024, 2025, 2026, 2027, 2028, 2029, 2030],
    'AI_Driven_Decision_Making': [10, 20, 35, 50, 70, 90, 100],
    'Decentralized_Web': [5, 15, 25, 40, 60, 80, 100]
}


df = pd.DataFrame(data)


plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['AI_Driven_Decision_Making'], marker='o', label='AI-Driven Decision Making', color='blue')
plt.plot(df['Year'], df['Decentralized_Web'], marker='o', label='Decentralized Web (Web 3.0)', color='green')


plt.title('Predicted Growth of AI-Driven Decision Making and Decentralized Web (2024-2030)', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Growth Percentage (%)', fontsize=12)
plt.xticks(df['Year'], rotation=45)
plt.yticks(range(0, 110, 10))
plt.grid(visible=True)


plt.legend()
plt.tight_layout()


plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

years = np.array([2024, 2026, 2028, 2030, 2032, 2034, 2036, 2038, 2040])
urban_population_percentage = np.array([50, 55, 60, 64, 68, 72, 74, 76, 78])
urbanization_data = pd.DataFrame({
    'Year': years,
    'Urban Population (%)': urban_population_percentage
})

plt.figure(figsize=(12, 7))
plt.plot(urbanization_data['Year'], urbanization_data['Urban Population (%)'], marker='o', linestyle='-', color='royalblue', linewidth=2, markersize=8)
plt.fill_between(urbanization_data['Year'], urbanization_data['Urban Population (%)'], color='lightblue', alpha=0.3)

for i in range(len(urbanization_data)):
    plt.text(urbanization_data['Year'][i], urbanization_data['Urban Population (%)'][i] + 0.5,
             f"{urbanization_data['Urban Population (%)'][i]}%",
             horizontalalignment='center', fontsize=10, color='black')

plt.title('Urbanization Trends Over the Years (2024-2040)', fontsize=16, fontweight='bold')
plt.xlabel('Year', fontsize=14)
plt.ylabel('Urban Population (%)', fontsize=14)
plt.xticks(years, fontsize=12)
plt.yticks(np.arange(50, 80, 5), fontsize=12)
plt.ylim(45, 80)
plt.grid(True, linestyle='--', alpha=0.7)
plt.axhline(y=70, color='orange', linestyle='--', label='Projected Urban Population by 2038')
plt.legend(fontsize=12)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

years = np.array([2023, 2024, 2025, 2026, 2027, 2028, 2029, 2030, 2031, 2032, 2033, 2034, 2035, 2036, 2037, 2038, 2039, 2040])
investment_amounts = np.array([1500, 1800, 2200, 2500, 2900, 3400, 3900, 4500, 5100, 5800, 6500, 7400, 8500, 9700, 11000, 12500, 14000, 16000])

investment_data = pd.DataFrame({
    'Year': years,
    'Investment (in millions $)': investment_amounts
})

plt.figure(figsize=(12, 7))
plt.plot(investment_data['Year'], investment_data['Investment (in millions $)'], marker='o', linestyle='-', color='mediumseagreen', linewidth=2, markersize=8)
plt.fill_between(investment_data['Year'], investment_data['Investment (in millions $)'], color='lightgreen', alpha=0.3)

for i in range(len(investment_data)):
    plt.text(investment_data['Year'][i], investment_data['Investment (in millions $)'][i] + 300,
             f"${investment_data['Investment (in millions $)'][i]}M",
             horizontalalignment='center', fontsize=10, color='black')

plt.title('Projected Investment in Neurotechnology (2023-2040)', fontsize=16, fontweight='bold')
plt.xlabel('Year', fontsize=14)
plt.ylabel('Investment (in millions $)', fontsize=14)
plt.xticks(years, fontsize=12)
plt.yticks(np.arange(0, 17000, 1000), fontsize=12)
plt.ylim(0, 17000)
plt.grid(True, linestyle='--', alpha=0.7)
plt.axhline(y=15000, color='orange', linestyle='--', label='Projected Investment in 2040')
plt.legend(fontsize=12)
plt.tight_layout()
plt.show()
```
